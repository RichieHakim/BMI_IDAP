```python
## Make convolution kernel

# First make a gaussian
gaussian_std_inSeconds = .1 # in seconds
kernel_duration = 91 # in S2p frames. Make sure it is an odd number

gaussian_std_inS2pFrames = gaussian_std_inSeconds * sample_rate_S2p
# kernel_shift = math.ceil(kernel_duration/2) # index of the zero point of the gaussian

kernel_gaussian = signal.gaussian(kernel_duration, gaussian_std_inS2pFrames)
kernel_gaussian = kernel_gaussian / np.sum(kernel_gaussian)
kernel_xAxis_inS2pFrames = np.arange(-(kernel_duration-1)/2, (kernel_duration-1)/2 +1)

plt.figure()
plt.plot(kernel_xAxis_inS2pFrames , kernel_gaussian)
print(kernel_xAxis_inS2pFrames)
```

```python
# do convolution

ws_rewards_bool_S2pInd_convolved = np.convolve(ws_rewards_bool_S2pInd , kernel_gaussian , mode='same')
ws_licks_bool_S2pInd_convolved = np.convolve(ws_licks_bool_S2pInd , kernel_gaussian , mode='same')
ws_treadmill_S2pInd_convolved = np.convolve(ws_treadmill_S2pInd , kernel_gaussian , mode='same')

dFoF_convolved = np.apply_along_axis(lambda m: np.convolve(m, kernel_gaussian, mode='same'), axis=1, arr=dFoF)
spks_convolved = np.apply_along_axis(lambda m: np.convolve(m, kernel_gaussian, mode='same'), axis=1, arr=spks)

body_part_highConfidence_interp_diff_S2pInd_conv = np.apply_along_axis(lambda m:
    np.convolve(m, kernel_gaussian, mode='same'), axis=0, arr=body_part_highConfidence_interp_diff_S2pInd)
body_part_highConfidence_interp_diff_abs_S2pInd_conv = np.apply_along_axis(lambda m:
    np.convolve(m, kernel_gaussian, mode='same'), axis=0, arr=np.abs(body_part_highConfidence_interp_diff_S2pInd))
body_part_highConfidence_interp_S2pInd_conv = np.apply_along_axis(lambda m:
    np.convolve(m, kernel_gaussian, mode='same'), axis=0, arr=np.abs(body_part_highConfidence_interp_S2pInd))

# check out result of convolution on ws data
fig = plt.figure()
ax1 = fig.add_subplot(311)
plt.plot(ws_rewards_bool_S2pInd_convolved , linewidth=0.5)
ax2 = fig.add_subplot(312)
plt.plot(ws_licks_bool_S2pInd_convolved , linewidth=0.5)
ax3 = fig.add_subplot(313)
plt.plot(ws_treadmill_S2pInd_convolved , linewidth=0.5)

# check out result of convolution on spks
plt.figure()
plt.plot(spks[3,:])
plt.plot(spks_convolved[3,:])

# check out result of convolution on DLC data
plt.figure()
plt.plot(body_part_highConfidence_interp_diff_S2pInd[:,2,0])
plt.plot(body_part_highConfidence_interp_diff_S2pInd_conv[:,2,0])
                                         
```

```python
# use the default value for reg_lambda
import pyglmnet

# glm = pyglmnet.GLM(distr='softplus', alpha=.1, score_metric='pseudo_R2', tol=1e-8,\
#                    reg_lambda=0.1, fit_intercept=True, max_iter = 10, solver='cdfast',\
#                    learning_rate=0.05, verbose=True)

glm = pyglmnet.GLMCV(distr='poisson', alpha=0.008, score_metric='pseudo_R2', tol=1e-8,\
                   reg_lambda=[0.2, 0.1, 0.05, 0.01], fit_intercept=True, max_iter = 1000, solver='cdfast',\
                   learning_rate=1e-3, verbose=True, cv=3)

# batch-gradient cdfast
# glm = pyglmnet.GLMCV(distr='softplus', alpha=0.01, score_metric='pseudo_R2',
#           learning_rate=0.05, tol=1e-8, verbose=True, max_iter=10000, cv=5, 
#                      reg_lambda=[0.4, 0.2, 0.1, 0.05])
# #                      reg_lambda=[0.4])

# fit model

glm.fit(X_train, y_train)

# score the test set prediction
y_test_hat = glm.predict(X_test)
y_train_hat = glm.predict(X_train)
y_hat = glm.predict(X)
print("test set pseudo $R^2$ = %f" % glm.score(X_test, y_test))

print(f'R_test:  {np.corrcoef(y_test, y_test_hat)[1,0]}')
print(f'R_train: {np.corrcoef(y_train, y_train_hat)[1,0]}')
print(f'beta_0:  {np.exp(glm.beta0_)}')
print(f'non-zero betas:  {np.sum(glm.beta_ !=0)}')
```
